```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("oregon_washington_combined_cleaned.csv", low_memory=False)
```

```python
df.columns[-100:]
```

```python
exclude_cols = ['License_Number', 'Hospital_Name', 'Year', 'Quarter', 'State']
numeric_cols = [col for col in df.columns if col not in exclude_cols and df[col].dtype in ['int64', 'float64']]

id_vars = ['Hospital_Name', 'Quarter', 'State']
melted = df.melt(id_vars = id_vars, value_vars = numeric_cols, var_name = 'metric', value_name = 'value')
grouped = melted.groupby(['Hospital_Name', 'Quarter', 'State', 'metric'])['value'].mean().reset_index()
grouped['metric_quarter'] = 'avg_' + grouped['metric'].str.lower() + '_q' + grouped['Quarter'].astype(str)

result = grouped.pivot_table(
    index = ['Hospital_Name', 'State'], 
    columns = 'metric_quarter', 
    values = 'value'
).reset_index()

result.columns.name = None
```

```python
result.head()
```

```python
result.to_csv('clustering_data.csv', index = False)
```
